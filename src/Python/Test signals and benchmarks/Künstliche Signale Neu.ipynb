```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
from math import inf
from numpy.linalg import norm
from sklearn.metrics import davies_bouldin_score as dbs
from sklearn.metrics import silhouette_score as ssc


def calc_centroids(arr, labels):
    singles = np.unique(labels)
    centroids = []
    j = 0
    k = 0
    for i in range(0, len(singles)):
        while j < len(labels) and labels[j] == singles[i]:
            j += 1
        sum = np.sum(arr[:, k:j], axis=1)
        sum = sum/(j + 1 - k)
        centroids.append(sum)
        k = j
    return centroids

def dunn_index(pf, cf, labels):
    """
    pf -- all data points
    cf -- cluster centroids
    """
    numerator = inf
    for i in range(0, len(cf)):        
        for j in range(0, len(cf)):            
            if i != j:
                dist = abs(norm(cf[i] - cf[j]))              
                if dist > 0 and dist < numerator:
                    numerator = dist
    
    denominator = 0
    k = 0
    for i in range(0, len(cf)):
        for j in range(0, len(pf)):
            if labels[j] == i + 1:
                for l in range(0, len(pf)):
                    if labels[l] == i + 1 and j != l:
                        denominator = max(denominator, abs(norm(pf[l] - pf[j])))
    
    return numerator/denominator

def index_I(arr, labels, centroids, p):
    singles = np.unique(labels)
    E_K = 0
    for i in range(0, len(singles)):
        for j in range(0, len(arr)):
            if i == 0 and labels[j] == singles[0]:
                E_1 = abs(norm(arr[j] - centroids[i]))
            if labels[j] == singles[i]:
                E = abs(norm(arr[j] - centroids[i]))
            E_K = E_K + E
    
    D_K = inf
    for i in range(0, len(singles)):
        for j in range(0, len(singles)):
            temp = abs(norm(centroids[i] - centroids[j]))
#             if temp > D_K:
#                 D_K = temp             
            if temp > 0 and temp < D_K:
                D_K = temp
    
    #index_I = ((1 / len(singles)) * (E_1 / E_K) * D_K) ** p
    index_I = (len(singles) * (D_K / E_K)) ** (1 / p)
    return(index_I)    


tot = np.loadtxt('outfile2.txt', dtype = int)
tot2 = []
for i in range(0, len(tot[0])):
    tot2.append(np.asarray([tot[0, i], tot[1, i], tot[2, i]]))

toople = []

for h in range(2, len(tot[0]) - 7):
    print("h:", h)
    print(datetime.datetime.now().time())
    print("")
    for i in range(h + 2, len(tot[0]) - 5):
        print("i:", i)
        print(datetime.datetime.now().time())
        print("")
        for j in range(i + 2, len(tot[0]) - 3):
            for k in range(j + 2, len(tot[0]) - 1):
                indices = [h, i, j, k]
                labels = []
                for l in range(0, h):
                    labels.append(1)
                for l in range(h, i):
                    labels.append(2)
                for l in range(i, j):
                    labels.append(3)
                for l in range(j, k):
                    labels.append(4)
                for l in range(k, len(tot[0])):
                    labels.append(5)
                
                centroids = calc_centroids(tot, labels)

                dav = dbs(np.transpose(tot), labels)
                dunn = dunn_index(tot2, centroids, labels)
                sil = ssc(np.transpose(tot), labels)
                index = index_I(tot2, labels, centroids, 2)

                toople.append([[dav, dunn, sil, index], [indices]])
                #print([[dav, dunn, sil, index], [indices]])

                
np.save('outfile4', toople)

# calibri = {'fontname':'Calibri'}
# plt.figure(figsize = (12, 8))
# plt.grid()
# plt.plot(tot, label = "Test signal") 
# plt.axvline(x = indices[0], color = "red", linestyle = 'dashed', label = "Segmentation line", ymin = 0.02, ymax = 0.98)
# for xc in range(1, len(indices)):
#     plt.axvline(x = indices[xc], color = "red", linestyle = 'dashed', ymin = 0.02, ymax = 0.98)
# plt.xlabel('Reading point / a.u.', **calibri, fontsize = 18)
# plt.ylabel('Signal range / a.u.', **calibri, fontsize = 18)
# plt.yticks(fontsize = 14)
# plt.xticks(fontsize = 14)
# plt.legend(fontsize = 12)
# plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


def metric(arr, seg):
    if len(seg) > 0:
        seg = np.sort(np.unique(seg))
        seg = seg - 1
        if seg[0] <= 0:
            return("Wrong segmentation. Can't use equal to or less than 1 as segmentation index.")
    if isinstance(arr, np.ndarray) == False:
        arr = np.array(arr)
        
    if len(arr.shape) == 1:
        if len(seg) <= 0:
            temp3 = (1 + 1/(len(arr))) / 2
            return(temp3)
        if seg[len(seg) - 1] > len(arr) - 1:
            return("Segmentation index or indices surpass array length.")
        
        temp = (max(arr[0:seg[0]]) - min(arr[0:seg[0]])) * seg[0]
        for i in range(0, len(seg) - 1):
            temp = temp + (max(arr[seg[i]:seg[i + 1]]) - min(arr[seg[i]:seg[i + 1]])) * (seg[i + 1] - seg[i])
        if seg[len(seg) - 1] == len(arr) - 1:
            temp = temp + 0
        else:
            temp = temp + (max(arr[seg[len(seg) - 1]:(len(arr) - 1)]) - min(arr[seg[len(seg) - 1]:(len(arr) - 1)])) * (len(arr) - seg[len(seg) - 1])

        temp2 = (max(arr) - min(arr)) * len(arr)

        temp3 = temp / temp2
        temp3 = temp3 + (len(seg) + 1) / len(arr)
        temp3 = temp3 / 2
        
    else:
        if len(seg) <= 0:
            temp3 = (1 + 1/(len(arr[0]))) / 2
            return(temp3)
        if seg[len(seg) - 1] > len(arr[0]) - 1:
            return("Segmentation index or indices surpass array length.")
        
        temp = 0
        for i in range(0, len(arr)):
            temp = temp + (max(arr[i][0:seg[0]]) - min(arr[i][0:seg[0]]))  * seg[0]
            for j in range(0, len(seg) - 1):
                temp = temp + (max(arr[i][seg[j]:seg[j + 1]]) - min(arr[i][seg[j]:seg[j + 1]])) * (seg[j + 1] - seg[j])
            if seg[len(seg) - 1] == len(arr[0]) - 1:
                temp = temp + 0
            else:
                temp = temp + (max(arr[i][seg[len(seg) - 1]:]) - min(arr[i][seg[len(seg) - 1]:])) * (len(arr[0]) - seg[len(seg) - 1])
        
        temp2 = 0
        for i in range(0, len(arr)):
            temp2 = temp2 + (max(arr[i]) - min(arr[i]))
        temp2 = temp2 * len(arr[0])

        temp3 = temp / temp2
        temp3 = temp3 + (len(seg) + 1) / len(arr[0])
        temp3 = temp3 / 2
        
    return(temp3)


A = np.loadtxt('outfile2.txt', dtype = int)
toople2 = []

for h in range(2, len(A[0]) - 3):
    for i in range(h + 1, len(A[0]) - 2):
        for j in range(i + 1, len(A[0]) - 1):
            for k in range(j + 1, len(A[0])):
                tempmet = metric(A, [h, i, j, k])
                toople2.append([[tempmet], [h - 1, i - 1, j - 1, k - 1]])

                
np.save('outfile6', toople2)
```

```python
import matplotlib.pyplot as plt
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

calibri = {'fontname':'Calibri'}
plt.figure(figsize = (12, 8))
plt.grid()
plt.scatter(x, y, color="blue")
plt.plot(x, mymodel, color="orange")
plt.xlabel('Variable 1 / a.u.', **calibri, fontsize = 18)
plt.ylabel('Variable 2 / a.u.', **calibri, fontsize = 18)
plt.yticks(fontsize = 14)
plt.xticks(fontsize = 14)
plt.savefig("linear_regression.svg", bbox_inches="tight")
plt.show()
```
