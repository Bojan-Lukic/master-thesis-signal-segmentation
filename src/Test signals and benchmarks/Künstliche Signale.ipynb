```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.ndimage.filters import gaussian_filter1d

# create equally spaced points between -10 and 10
x = np.linspace(-5, 5, 41)

# calculate the y value for each element of the x vector
y = -x**2 + 50
y2 = np.zeros(30) + 20
y3 = y2
y3 = np.append(y3, y)
y3 = np.append(y3, y2) 

ysmoothed = gaussian_filter1d(y3, sigma=4)
np.savetxt('outfile.txt', ysmoothed, fmt='%d')

calibri = {'fontname':'Calibri'}
plt.figure(figsize = (12, 8))
plt.grid()
plt.plot(ysmoothed)    
plt.xlabel('Reading point r', **calibri, fontsize = 18)
plt.ylabel('Signal range', **calibri, fontsize = 18)
plt.yticks(fontsize = 14)
plt.xticks(fontsize = 14)

plt.savefig("test_signal.svg", bbox_inches="tight")
plt.show()

# fig, ax = plt.subplots()
# ax.plot(x, y)
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
import numpy as np
import pandas as pd


tot = np.loadtxt('outfile2.txt', dtype = int)

for i in range(0, 3):
    calibri = {'fontname':'Calibri'}
    ax = plt.figure(figsize = (12, 8)).gca()
    plt.grid()
    plt.plot(tot[i])
    plt.xlabel('Reading point r', **calibri, fontsize=18)
    plt.ylabel('Signal range', **calibri, fontsize=18)
    plt.yticks(fontsize = 14)
    plt.xticks(fontsize = 14)
    ax.xaxis.set_major_locator(MaxNLocator(integer=True))
    plt.show()


fig, (ax1, ax2, ax3) = plt.subplots(3, figsize=(12, 8))
ax1.grid()
ax2.grid()
ax3.grid()
ax1.plot(tot[0])
ax2.plot(tot[1], color="orange")
ax3.plot(tot[2], color="green")
plt.xlabel('Reading point r', fontsize=14)
ax1.xaxis.set_major_locator(MaxNLocator(integer=True))
ax2.xaxis.set_major_locator(MaxNLocator(integer=True))
ax3.xaxis.set_major_locator(MaxNLocator(integer=True))
ax2.set_ylabel('Signal range', fontsize=14)

plt.savefig("multivariate_test_signal.svg", bbox_inches="tight")
plt.show()
```
